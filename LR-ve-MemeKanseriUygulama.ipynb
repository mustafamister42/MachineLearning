```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#Veri Seti Oluşturma

```python
X,y=make_classification(n_samples=1000,n_features=10,n_informative=2,n_classes=2,random_state=42)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

#Model Eğitimi

```python
lr_model=LogisticRegression()
lr_model.fit(X_train,y_train)
```

```python
y_pred=lr_model.predict(X_test)
basari_skoru=accuracy_score(y_test,y_pred)
print("Başarı Skoru:",basari_skoru)
```

#KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=6)
knn_model.fit(X_train,y_train)
y_pred=knn_model.predict(X_test)
basari=accuracy_score(y_test,y_pred)
print("Başarı Skoru:",basari)
```

#SVR

```python
from sklearn.svm import SVC
svr_model=SVC(kernel='linear')
svr_model.fit(X_train,y_train)
y_pred=svr_model.predict(X_test)
basari=accuracy_score(y_test,y_pred)
print("Başarı Skoru:",basari)
```

```python
from google.colab import drive
import pandas as pd
drive.mount('/content/drive')
veri_yol='/content/drive/MyDrive/Colab Notebooks/memekanseri.data'
veri=pd.read_csv(veri_yol,header=None)
kolon_isimleri=veri.iloc[0]
print(kolon_isimleri)
print(veri.head())
```

```python
veri.columns=["Yas","Ameliyat_yili","Bulgular","Canli"]
print(veri.head(10))
```

```python
veri.loc[veri["Canli"]==2,"Canli"]=0
print(veri.head(10))
```

```python
print(veri.dtypes)
```

```python
veri["Canli"]=veri["Canli"].astype("int64")
print(veri.dtypes)
```

```python
from sklearn.preprocessing import MinMaxScaler
X=veri.drop("Canli",axis=1)
y=veri["Canli"]
olceklendirme=MinMaxScaler()
X=olceklendirme.fit_transform(X)
X=pd.DataFrame(X,columns=["Yas","Ameliyat_yili","Bulgular"])
print(X.head())
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=34)
```

```python
from keras.models import Sequential
from keras.layers import Dense
ysa_model=Sequential()
ysa_model.add(Dense(128,input_shape=(X_train.shape[1],),activation="relu"))#Giriş Katmanı
ysa_model.add(Dense(64,activation="relu"))#Gizli katman1
ysa_model.add(Dense(32,activation="relu"))#Gizli katman2
ysa_model.add(Dense(16,activation="relu"))#Gizli katman3
ysa_model.add(Dense(1,activation="sigmoid"))#Çıkış Katmanı
ysa_model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
ysa_model.fit(X_train,y_train,epochs=1000,verbose=0)
```

```python
kayıp,basari=ysa_model.evaluate(X_test,y_test)
print("Kayıp:",kayıp)
print("Başarı:",basari)
```
