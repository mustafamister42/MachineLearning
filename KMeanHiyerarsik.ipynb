```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
X,_=make_blobs(n_samples=300 #300 veri noktası oluşturur
               ,centers=4,   #4 farklı küme(cluster) olsun
               random_state=55)#Rastgelelik kontrolü için sabit seed değeri

```

```python
kmeans=KMeans(n_clusters=4)
kmeans.fit(X)
```

#Hiyerarşik Kümeleme

```python
from scipy.cluster.hierarchy import dendrogram,linkage

hiyerarsik=linkage(X,"single")
#hiyerarsik=linkage(X,"complete")
plt.figure(figsize=(10,7))
plt.title("Hiyerarşik Kümeleme Dendrogramı")
dendrogram(hiyerarsik)
plt.show()

```
